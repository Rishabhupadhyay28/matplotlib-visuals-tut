```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
m=np.linspace(0,10,11)
```

```python
m
```

```python
#speed of light
c=3*10**8
```

```python
E=m*c**2
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E = mc^2')
plt.xlabel('M in g')
plt.ylabel('E in J ')
plt.xlim(0,10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=5)
ax.set_xlim(0,10)
```

# logrithmic scale 

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E = mc^2')
plt.xlabel('M in g')
plt.ylabel('E in J ')
plt.yscale('log')
plt.grid( axis='y',which='both')#which='both')
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
plt.legend()    #ax.legend()
```

```python
fig=plt.figure(dpi=200)
ax=fig.add_axes([0,0,1,1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
plt.legend(loc=((1.05,0.5)) )  #ax.legend()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(12,8))

axes[0].plot(labels,july16_2007)
axes[0].set_title('JULY16_2007')

axes[1].plot(labels,july16_2020)
axes[1].set_title('JULY16_2020')
```

```python
fig,ax1=plt.subplots(figsize=(12,8))

ax1.plot(labels,july16_2007,lw=2,color="blue")
ax1.set_ylabel("July16_2007",fontsize=18,color='blue')
ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(3)

for label in ax1.get_yticklabels():
    label.set_color('blue')
    
ax2=ax1.twinx()
ax2.plot(labels,july16_2020,lw=2,color="red")
ax2.set_ylabel("July16_2007",fontsize=18,color='red')
ax1.spines['right'].set_color('red')
ax1.spines['right'].set_linewidth(3)

for label in ax2.get_yticklabels():
    label.set_color('red')
```
