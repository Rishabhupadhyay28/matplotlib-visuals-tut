```python
import matplotlib.pyplot as plt
import numpy as np
```

# OOP matplotlib

```python
x=np.arange(0,10)
y=x*2

a=np.linspace(0,10,11)
b=a**4
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

inner graph

```python
fig=plt.figure(dpi=200)

#large axes

axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_ylabel('A  y-axis ')
axes1.set_xlabel('B x-axis')
axes1.set_title('this is my title')

#small axes
axes2=fig.add_axes([0.2,0.7,0.2,0.2])#(x-axis,y-axis,height,width)
axes2.plot(x,y)

axes2.set_xlim(1,2.5)
axes2.set_ylim(0,20)
axes2.set_xlabel('x-axis g-2')
axes2.set_ylabel('y-axis g-2')
axes2.set_title('title 2 graph')
plt.show()
```

# SUBPLOT

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[1,0].plot(a,b)
axes[1,1].plot(x,y)
axes[0,1].plot(a,b)
fig.subplots_adjust(wspace=True,hspace=True)
```

```python
import matplotlib.pyplot as plt
x = np.arange(0, 3.14*2, 0.05)
y = np.sin(x)
plt.plot(x,y,'r:')
plt.title("this is my title one",color='red',size=10)
plt.plot()
```

```python
x=np.linspace(0,10,11)

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2 ,label="x vs x**2")
ax.legend(loc='upper center')
ax.legend(loc=[0.8,0.8])
```

# Matplotlib styling

```python
x=np.linspace(0,11,10)
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
#axes.plot(x,x,color='yellow', lw=5,ls='--') #grb() HEX code
axes.plot(x,x+1,color='purple',linewidth=5,ls=':')
```
